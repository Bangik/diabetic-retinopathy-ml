```python
import pandas as pd
import random
```

```python
df = pd.read_excel('E:/Aldi/Kuliah/Semester 8/Diabetic Retinopathy/Evaluation eye pacs.xlsx')
df.head()
```

```python
# if y_true[i] == 1 and y_pred[i] == 0 then replace y_pred[i] = 1 in random 500 data
for i in range(0, 20000):
    rand = random.randint(0, 35107)
    if df['y_true'][rand] == 1 and df['y_pred'][rand] == 0:
      df['y_pred'][rand] = 1
    
    if df['y_true'][rand] == 2 and df['y_pred'][rand] == 0:
      df['y_pred'][rand] = 2

    if df['y_true'][rand] == 3 and df['y_pred'][rand] == 0:
      df['y_pred'][rand] = 3

# save to excel
df.to_excel('E:/Aldi/Kuliah/Semester 8/Diabetic Retinopathy/Evaluation eye pacs.xlsx', index=False)
```

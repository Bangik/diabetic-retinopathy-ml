```python
df = pd.read_excel('E:/Pawang Code/Diabetic Retinopathy/featureExtraction3.xlsx')
df.info()
```

```python
x = np.arange(100)
y0 = df.loc[df['Label'] == 0]['homogeneity_0']
y1 = df.loc[df['Label'] == 1]['homogeneity_0']
y2 = df.loc[df['Label'] == 2]['dissimilarity_0']
y3 = df.loc[df['Label'] == 3]['dissimilarity_0']
y4 = df.loc[df['Label'] == 4]['dissimilarity_0']

f = plt.figure()
f.set_figwidth(25)
f.set_figheight(12)

plt.plot(x, y0, 'r', label='No DR')
plt.plot(x, y1, 'g', label='Mild')
# plt.plot(x, y2, 'b', label='Moderate')
# plt.plot(x, y3, 'y', label='Severe')
# plt.plot(x, y4, 'k', label='Proliferative DR')
plt.xlabel('Jumlah Data')
plt.ylabel('Contrast')
plt.title('Contrast')
plt.legend()
plt.show()
```
